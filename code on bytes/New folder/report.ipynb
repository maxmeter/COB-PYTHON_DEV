```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('expenses - Sheet2.csv', parse_dates=['Date'])

df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year


unique_months_years = df[['Month', 'Year']].drop_duplicates()

for index, row in unique_months_years.iterrows():
    month = row['Month']
    year = row['Year']

    monthly_expenses = df[(df['Month'] == month) & (df['Year'] == year)]

    total_expenses = monthly_expenses['Amount'].sum()

    category_expenses = monthly_expenses.groupby('Category')['Amount'].sum()
    category_expenses.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Monthly Expenses for {month}/{year} by Category')
    plt.show()

    print(f'Total expenses for {month}/{year}: {total_expenses:.2f}')
```
